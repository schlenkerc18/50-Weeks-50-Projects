# Modeling

```python
predictors = ['age', 'bmi', 'smoker']
target = 'charges'

X_test, X_train, y_test, y_train = train_test_split(df[predictors], df[target], train_size = .8, random_state=42)

xgb = XGBRegressor()
xgb.fit(X_test, y_test)
```

```python
df_test = pd.merge(X_test, y_test, how = 'inner', left_index = True, right_index = True)
df_test['Prediction'] = xgb.predict(df_test[predictors])
df_test.head()
```

```python
untuned_error = mean_absolute_error(df_test.charges, df_test.Prediction)
print("Average Absolute error:", "${:.2f}".format(untuned_error))
```

# Using Grid Search to Tune Hyperparameters

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [.01, .05, .1],
    'max_depth': [3,5,7,9],
}
```

```python
xgb_grid = GridSearchCV(xgb, param_grid, scoring = 'neg_mean_absolute_error', cv = 10)
xgb_grid.fit(X_test, y_test)
```

```python
best_params = xgb_grid.best_params_
best_params
```

```python
X_test, X_train, y_test, y_train = train_test_split(df[predictors], df[target], train_size = .8, random_state=42)

xgb = XGBRegressor(n_estimators = best_params['n_estimators'], max_depth = best_params['max_depth'],
                   learning_rate = best_params['learning_rate'])
xgb.fit(X_test, y_test)
```

```python
df_test = pd.merge(X_test, y_test, how = 'inner', left_index = True, right_index = True)
df_test['TunedPrediction'] = xgb.predict(df_test[predictors])
df_test.head()
```

```python
tuned_error = mean_absolute_error(df_test.charges, df_test.TunedPrediction)

print("Avg. Abs. Error for untuned model:", "${:.2f}".format(untuned_error))
print("Avg. Abs. Error for tuned model:", "${:.2f}".format(tuned_error))
```

# Modeling with Tensorflow

```python
X_test, X_train, y_test, y_train = train_test_split(df[predictors], df[target], train_size = .8, random_state=42)

X_train = np.array(X_train)
y_train = np.array(y_train).reshape(-1,1)

feature_normalizer = tf.keras.layers.Normalization(axis = -1)
feature_normalizer.adapt(X_train)

X_train_norm = feature_normalizer(X_train)
```

```python
X_train
```

```python
# define the model where the number of units represents the number of neurons in each layer

model = Sequential([
    feature_normalizer, # normalization layer
    Dense(units = 50, activation = 'linear', kernel_regularizer = tf.keras.regularizers.l2(.01)),
    Dense(units = 25, activation = 'linear', kernel_regularizer = tf.keras.regularizers.l2(.01)),
    Dense(units = 1, activation = 'linear', kernel_regularizer = tf.keras.regularizers.l2(.01))
])
```

```python
# define loss function
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate = 0.1),
    loss = 'mean_absolute_error'
)
```

```python
X_train
```

```python
# train model

history = model.fit(
    X_train,
    y_train,
    epochs = 1000, # iterations
    verbose = 0, # suppress logging
    validation_split = .2 # calculate validation results on 20% of data
)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')

plt.title("Line plot of MAE")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend()
plt.grid(True)
```

```python
df_test = pd.merge(X_test, y_test, how = 'inner', left_index = True, right_index = True)
X_test_norm = feature_normalizer(X_test)
df_test['NeuralNetPred'] = model.predict(X_test_norm)

df_test
```
