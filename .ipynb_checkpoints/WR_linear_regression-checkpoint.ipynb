### Anova (Draft Capital and Exp)

```python
def perm_test(df, col):
    df = df.copy()
    df[col] = np.random.permutation(df[col].values)
    return df.groupby(col).aggregate({'rec_yards_gained': np.mean}).var()[0]
```

#### Draft Capital Anova

```python
draft_cap_var = df.groupby('draft_capital').aggregate({'rec_yards_gained': np.mean}).var()[0]
df.groupby('draft_capital').aggregate({'rec_yards_gained': np.mean})
```

```python
print("observed means:", df.groupby('draft_capital').aggregate({'rec_yards_gained': np.mean}).values.ravel())
print("Observed variance:", draft_cap_var)
```

```python
draft_capital_var = [perm_test(df, 'draft_capital') for _ in range (3000)]
```

```python
# the probability of this happening by chance is essentially 0
print("Pr(Prob):", np.mean([var > draft_cap_var for var in draft_capital_var]))
```

#### Exp Anova

```python
exp_var = df.groupby('exp_class').aggregate({'rec_yards_gained': np.mean}).var()[0]
df.groupby('exp_class').aggregate({'rec_yards_gained': np.mean})
```

```python
print("observed mean:", df.groupby('exp_class').aggregate({'rec_yards_gained': np.mean}).values.ravel())
print("observed variance:", exp_var)
```

### Linear Regression

```python
def compute_loss(X, y, w, b):
    """
    args:
        X: scalar vector of input features
        y: scalar vector of target var

        w: scalar parameter of model
        b: scalar parameter of model
    
    returns:
        total loss of the model
    """
    
    total_loss = 0
    m = X.shape[0] # number of records
    
    for i in range(m):
        y_hat = w * X[i] + b
        
        error = (y_hat - y[i]) ** 2
        total_loss += error
        
    return total_loss / (2 * m)
```

```python
def compute_gradient(X, y, w, b):
    """
    args: 
        X: scalar vector of input features
        y: scalar vector of target var

        w: scalar parameter of model
        b: scalar parameter of model
        
    returns:
        gradient at model paramters w, b
    """
    
    dj_dw = 0
    dj_db = 0
    
    m = X.shape[0]
    # f_wb = w * x[i] + b
    # cost function -> sum of: (f_wb - y) ** 2
    # gradient calculates partial derivative of cost function
    # what is partial derivative 
    for i in range(m):
        f_wb = w * X[i] + b
        
        dj_dw_i = (f_wb - y[i]) * X[i]
        dj_db_i = f_wb - y[i]
        
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, iterations):
    """
    args:
        X: feature vector
        y: target vector
        w,b: scalar parameters for model
    
    returns:
        optimized parameters w,b
    """
    
    w, b = w_in, b_in

    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(X, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        cost = compute_loss(X, y, w, b)
        
        if (i % math.ceil(iterations / 10) == 0):
            print("w: " + str(w) + ", b: " + str(b) + ", total_cost:" + str(cost))
            
    return w, b
```

```python
X_train = np.array(df['receiving_air_yards']).reshape(df.shape[0], -1)
y = np.array(df['rec_yards_gained']).reshape(df.shape[0], -1)

alpha = .01
iterations = 10000

w = 10
b = 5

w_out, b_out = gradient_descent(X_train, y, w, b, alpha, iterations)
```

```python
plt.figure()

sns.scatterplot(data = df, x = 'receiving_air_yards', y = 'rec_yards_gained')
y_hat = w_out * X_train + b_out
plt.plot(X_train, y_hat, color = 'red')


plt.title("Linear Regression Plot")
plt.show()
```

### Multiple Linear Regression

```python
def compute_ml_cost(x, y, w, b):
    """
    args:
        x: vector of feature variables
        y: vector of feature variables
        w: vector of scalars
        b: scalar parameter
        
    output:
        total cost associated with model parameters w,b
    """
    
    m = x.shape[0]
    total_cost = 0
    
    for i in range(m):
        f_wb = np.dot(w,x[i]) + b
        error = (f_wb - y[i]) ** 2
        total_cost += error
        
    total_cost = total_cost / (2 * m)
    
    return total_cost 
```

```python
def compute_ml_gradient(x, y, w, b):
    """
    args:
        x: vector of feature variables
        y: vector of feature variables
        w: vector of scalars
        b: scalar parameter
    """
    
    m,n = x.shape
    
    dj_dw = np.zeros((n,))
    dj_db = 0
    
    for i in range(m):
        f_wb = np.dot(w, x[i]) + b
        
        for j in range(n):
            dj_dw[j] = dj_dw[j] + ((f_wb - y[i]) * x[i, j])
        
        dj_db += f_wb - y[i]
        
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw, dj_db
```

```python
def gradient_descent_ml(x, y, w_in, b_in, alpha, iterations):
    """
    args:
        x: 2d vector of feature variables
        y: vector of target variable
        w_in: parameter vector
        b_in: scalar parameter
        alpha: scalar learning rate
        iterations: num of iterations for gd alg
        
    output:
        w,b: parameters for model
        cost: total cost for w,b
    """
    J_history = []
    w = copy.deepcopy(w_in)
    b = b_in
    
    for i in range(iterations):
        dj_dw, dj_db = compute_ml_gradient(x, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
                # Save cost J at each iteration
        if (i<100000):      # prevent resource exhaustion 
            J_history.append( compute_ml_cost(x, y, w, b))        
        
        if (i% math.ceil(iterations / 10) == 0):
            print(f"Iteration {i}: Cost {J_history[-1]}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```

```python
predictors = ['receiving_air_yards', 'Round']
outcome = 'rec_yards_gained'

X_train = np.array(df[predictors])
y = np.array(df[outcome])

alpha = .01
iterations = 10000

w = [1, 2]
b = 1

w_out, b_out, _ = gradient_descent_ml(X_train, y, w, b, alpha, iterations)
```
