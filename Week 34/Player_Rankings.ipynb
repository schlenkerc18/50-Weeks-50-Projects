```python
import pandas as pd
import numpy as np
```

```python
qb_df = pd.read_excel('qb_ratings.xlsx')
rb_df = pd.read_excel('rb_ratings.xlsx')
wr_df = pd.read_excel('wr_ratings.xlsx')
te_df = pd.read_excel('te_ratings.xlsx')

qb_df['Position'] = 'QB'
rb_df['Position'] = 'RB'
wr_df['Position'] = 'WR'
te_df['Position'] = 'TE'
```

```python
qb_df = qb_df[['Player', 'Position', 'Rating']].copy()

qb_df.sort_values(by = 'Rating', inplace = True, ascending = False)
qb_df.reset_index(inplace=True)
qb_df.rename(columns = {'index': 'PosRank'}, inplace=True)
qb_df.PosRank = qb_df.PosRank + 1

# push qb rating down
qb_df.Rating = qb_df.Rating - 200
qb_df.head()
```

```python
rb_df = rb_df[['Player', 'Position', 'Rating']].copy()

rb_df.sort_values(by = 'Rating', inplace = True, ascending = False)
rb_df.reset_index(inplace=True)
rb_df.rename(columns = {'index': 'PosRank'}, inplace = True)
rb_df.PosRank = rb_df.PosRank + 1
rb_df.head()
```

```python
wr_df = wr_df[['Player', 'Position', 'Rating']].copy()


wr_df.sort_values(by = 'Rating', inplace = True, ascending = False)
wr_df.reset_index(inplace = True)
wr_df.rename(columns = {'index': 'PosRank'}, inplace = True)
wr_df.PosRank = wr_df.PosRank + 1
wr_df.head(10)
```

```python
te_df = te_df[['Player', 'Position', 'Rating']]

te_df.sort_values(by = 'Rating', inplace = True, ascending = False)
te_df.reset_index(inplace = True)
te_df.rename(columns = {'index': 'PosRank'}, inplace = True)
te_df.PosRank = te_df.PosRank + 1

# push te_rankings down
te_df.Rating = te_df.Rating - 100
te_df.head()
```

```python
df = pd.concat([qb_df, rb_df, wr_df, te_df], ignore_index = True)
df
```

```python
new_df = df.sort_values(by='Rating', ascending=False, ignore_index = True).reset_index(drop = False)\
                                                                      .rename(columns={'index': 'Rank'})
new_df.Rank = new_df.Rank + 1
new_df.rename(columns = {'Rank': 'OverallRank'}, inplace = True)
new_df = new_df[['OverallRank', 'Player', 'Position', 'PosRank', 'Rating']]
new_df.to_csv('player_rankings.csv', encoding='utf-8', index = False)
new_df.to_csv('C:/Users/schle/PycharmProjects/FantasyFootballDashboard/StatFiles/player_rankings.csv',
              encoding = 'utf-8',
              index = False)
```
