### Permutation Tests

```python
def perm_fun(df, len_a, col):
    a = df.sample(len_a, replace = False)
    b = df[~df.index.isin(a.index)]
    
    return a[col].mean() - b[col].mean()
```

```python
count = df.groupby('class').aggregate({'class': 'count'})
count
```

```python
malignant_count = count.loc[4]['class']
malignant_count
```

#### Clump Thickness

```python
actual_diff = df_group.loc[4].clump_thickness - df_group.loc[2].clump_thickness
diffs = [perm_fun(df, malignant_count, 'clump_thickness') for _ in range(1000)]
print(actual_diff)
```

```python
plt.figure(figsize = (6,4))
sns.histplot(diffs)
plt.axvline(x = actual_diff, color = 'red')
plt.show()
```

```python
p_val = np.array([1 if np.abs(diff) > actual_diff else 0 for diff in diffs])
p_val.mean()
```

### Modeling

```python
# using all variables as predictors since they are all statistically significant
df.columns
```

```python
predictors = ['clump_thickness', 'uniformity_of_cell_size',
       'uniformity_of_cell_shape', 'marginal_adhesion',
       'single_epithelial_cell_size', 'bare_nuclei', 'bland_chromatin',
       'normal_nucleoli', 'mitosis']
outcome = 'class'

X = df[predictors]
y = df[outcome]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3)

rf = RandomForestClassifier(n_estimators = 500, min_samples_split = 10)
rf.fit(X_train, y_train)
```

```python
df['RFPrediction'] = rf.predict(df[predictors])
df.head()
```

```python
print("Accuracy:" , np.round(accuracy_score(df['RFPrediction'], df['class']), 3))
print("Precision:", np.round(precision_score(df['RFPrediction'], df['class'], pos_label = 4), 3))
print("Recall:", np.round(recall_score(df['RFPrediction'], df['class'], pos_label = 4), 3))
```
