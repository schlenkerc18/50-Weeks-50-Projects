```python
import pandas as pd
import numpy as np

import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```python
drafted_receivers = pd.read_csv('./data/wr_draft_position.csv')
drafted_receivers.head()
```

```python
df_2022 = pd.read_csv('./data/sportsref_download_2022.csv')

df_2022.Player = df_2022.Player.str.strip('*')
df_2022.fillna(0, inplace = True)
df_2022.rename(columns={'Player': 'Name'}, inplace = True)
df_2022['Year'] = 2022

df_2022 = df_2022[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2022.sort_values('Yds', ascending = False, inplace = True)
df_2022.head()
```

```python
df_2021 = pd.read_csv('./data/sportsref_2021.csv')

df_2021.Player = df_2021.Player.str.strip('*')
df_2021.fillna(0, inplace = True)
df_2021.rename(columns={'Player': 'Name'}, inplace = True)
df_2021['Year'] = 2021

df_2021 = df_2021[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2021.head()
```

```python
df_2020 = df_2020 = pd.read_csv('./data/sportsref_download_2020.csv')

df_2020.Player = df_2020.Player.str.strip('*')
df_2020.fillna(0, inplace = True)
df_2020.rename(columns={'Player': 'Name'}, inplace = True)
df_2020['Year'] = 2020

df_2020 = df_2020[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2020.head()
```

```python
df_2019 = df_2019 = pd.read_csv('./data/sportsref_download_2019.csv')

df_2019.Player = df_2019.Player.str.strip('*')
df_2019.fillna(0, inplace = True)
df_2019.rename(columns={'Player': 'Name'}, inplace = True)
df_2019['Year'] = 2019

df_2019 = df_2019[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2019.head()
```

```python
df_2018 = pd.read_csv('./data/sportsref_download_2018.csv')

df_2018.Player = df_2018.Player.str.strip('*')
df_2018.fillna(0, inplace = True)
df_2018.rename(columns={'Player': 'Name'}, inplace = True)
df_2018['Year'] = 2018

df_2018 = df_2018[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2018.head()
```

```python
df_2017 = pd.read_csv('./data/sportsref_download_2017.csv')

df_2017.Player = df_2017.Player.str.strip('*')
df_2017.fillna(0, inplace = True)
df_2017.rename(columns={'Player': 'Name'}, inplace = True)
df_2017['Year'] = 2017

df_2017 = df_2017[['Name', 'School', 'Conf', 'G', 'Rec', 'Yds', 'Avg', 'TD', 'Year']]
df_2017.head()
```

```python
concat_df = pd.concat([df_2017, df_2018, df_2019, df_2020, df_2021], axis = 0)

# sort be Year descending so that we can drop duplicates
# we only want receivers final year of production in college
concat_df.sort_values('Year', inplace = True, ascending = False)
concat_df.drop_duplicates('Name', inplace = True)

concat_df.head()
```

```python
concat_df.Year.value_counts()
```

```python
merged_df = pd.merge(concat_df, drafted_receivers, how = 'inner', on = 'Name')

# get z-scores
# merged_df['Yds'] = (merged_df.Yds - merged_df.Yds.mean()) / merged_df.Yds.std()
# merged_df['TD'] = (merged_df.TD - merged_df.TD.mean()) / merged_df.TD.std()

merged_df.head()
```

```python
predictors = ['Yds', 'TD']
outcome = 'Round'

new_player = df_2021.loc[df_2021.Name == 'Jaxon Smith-Njigba', predictors]

X = merged_df[predictors]
y = merged_df[outcome]

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X,y)
knn.predict(new_player)
```

```python
# knn algorithm is predicting round 1 for Quentin Johnston
nbrs = knn.kneighbors(new_player)
X.iloc[nbrs[1][0], :]
```

```python
# Scaling first
new_player = df_2021.loc[df_2021.Name == 'Jaxon Smith-Njigba', predictors]

X = merged_df[predictors]
y = merged_df[outcome]

scaler = preprocessing.StandardScaler()
scaler.fit(X * 1.0)

X_std = scaler.transform(X * 1.0)
new_player_std = scaler.transform(new_player * 1.0)

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_std, y)

nbrs = knn.kneighbors(new_player_std)
X.iloc[nbrs[1][0], :]
```
