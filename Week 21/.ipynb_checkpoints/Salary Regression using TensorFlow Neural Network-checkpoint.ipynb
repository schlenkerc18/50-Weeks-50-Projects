```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('Salary_Data.csv')

df.columns = ['age', 'gender', 'educationLevel', 'jobTitle', 'yearsOfExp', 'salary']
df.dropna(axis = 0, inplace = True)

df.head()
```

```python
df.groupby('gender').aggregate({'salary': 'mean'})
```

```python
df.isna().sum()
```

```python
len(df.jobTitle.unique())
```

```python
job_group = df.groupby('jobTitle').aggregate({'salary': np.average, 
                                  'educationLevel': 'count'}).rename(columns = {'educationLevel': 'Total'})
job_group = job_group[job_group.Total >= 25]
job_group.reset_index(drop = False, inplace = True)
job_group.sort_values(by = 'salary', ascending = False)[0:10]
```

### Exploring Data through Visualization

```python
# top 10 jobs by salary

plt.figure(figsize = (6,4))

sns.barplot(data = job_group.sort_values('salary', ascending = False)[0:10],
            x = 'salary', y = 'jobTitle', palette = 'Paired')

plt.title("Top 10 Highest Paying Positions")
plt.ylabel('')
plt.xlabel('Salary in USD')
plt.show()
```

```python
g = sns.jointplot(data = df, x = 'age', y = 'salary', hue = 'gender', palette = 'muted')
g.ax_marg_x.set_xlim(15, 70)
plt.show()
```

```python
g = sns.jointplot(data = df, x = 'yearsOfExp', y = 'salary', hue = 'educationLevel', palette = 'muted')
g.ax_marg_x.set_xlim(0, 40)
g.ax_marg_y.set_ylim(0, 300000)
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(df.corr().round(2), annot = True)
plt.show()
```
