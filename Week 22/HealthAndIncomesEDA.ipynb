# TensorFlow Model

> Let's Train a model to predict a country's average life expectancy!<br>
>> Input Variable: GDP

```python
predictors = ['infant_mortality', 'gdp']
target = 'life_expectancy'

X = np.array(df[predictors], ndmin = 2)
y = np.array(df[target]).reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = .8)

norm_layer = tf.keras.layers.Normalization(axis = -1)
norm_layer.adapt(X_train) # learns mean and variance
X_train_norm = norm_layer(X_train)
```

# Define Model

```python
tf.random.set_seed(1234)

# define model with 3 layers
# 1st layer has 5 neurons and uses linear activation function
# output layer has 1 neuron and uses linear activation function

model = Sequential(
    [
        Dense(5, activation = 'relu', name = 'layer_1'),
        Dense(1, activation = 'linear', name = 'output_layer')
    ]
)
```

# Define Loss Function

```python
# define loss

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
    loss = 'mean_absolute_error'
)
```

# Train Model

```python
history = model.fit(X_train_norm, y_train, 
          epochs = 100, 
          # suppress logging
          verbose = 0,
          # calculate validation results on 20% of the training data
          validation_split = 0.2)
```

# Checking For Variance

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.ylim([0, 10])

plt.title("Line plot of MAE")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend()
plt.grid(True)
```

```python
X_test_norm = norm_layer(X_test)

predictions = model.predict(X_test_norm)
```

# Preview of Predictions

```python
x = pd.DataFrame(data = X_test, columns = ['Infant_Mortality', 'GDPinBillions'])
y = pd.DataFrame(data = y_test, columns = ['LifeExpectancy'])

df_test = pd.merge(x, y, how = 'left', left_index = True, right_index = True)
df_test['Pred'] = predictions

df_test.head(20)
```
